```python
import pandas as pd
```

```python
inp = pd.read_csv('files/inputevents_mv.csv')
```

```python
inp['STARTTIME'] = pd.to_datetime(inp['STARTTIME'])
inp['ENDTIME'] = pd.to_datetime(inp['ENDTIME'])
inp['drug_length'] = (inp['ENDTIME'] - inp['STARTTIME']).dt.total_seconds()/86400
inp['drug_length'].head(20)
```

```python
#df.drop(columns=['ROW_ID','SUBJECT_ID','ICUSTAY_ID','STARTTIME','ENDTIME','RATE','RATEUOM','STORETIME','CGID',''], inplace=True)
```

```python
#inp['ORDERCATEGORYNAME'].nunique()
#inp.ORDERCATEGORYNAME.unique()
#inp[inp['AMOUNTUOM']=='mEq.']
inp = inp[['HADM_ID','ITEMID','AMOUNT','AMOUNTUOM','PATIENTWEIGHT','ORDERID','drug_length']]
```

```python
inp['AMOUNTUOM'].replace('mEq','1', inplace=True)
inp['AMOUNTUOM'].replace('ml','2', inplace=True)
inp['AMOUNTUOM'].replace('dose','3', inplace=True)
inp['AMOUNTUOM'].replace('grams','4', inplace=True)
inp['AMOUNTUOM'].replace('units','5', inplace=True)
inp['AMOUNTUOM'].replace('mg','6', inplace=True)
inp['AMOUNTUOM'].replace('mcg','7', inplace=True)
inp['AMOUNTUOM'].replace('mmol','8', inplace=True)
inp['AMOUNTUOM'].replace('L','9', inplace=True)
inp['AMOUNTUOM'].replace('mEq.','10', inplace=True)
inp['AMOUNTUOM'].replace('ml/hr','11', inplace=True)
inp['AMOUNTUOM'].replace('International Units','12', inplace=True)
inp['AMOUNTUOM'].replace('uL','13', inplace=True)
inp['AMOUNTUOM'].replace('ounces','14', inplace=True)
inp['AMOUNTUOM'].replace('pg','15', inplace=True)
inp['AMOUNTUOM'].replace('/hour','16', inplace=True)
```

```python
inp
```

```python
df_group = inp.groupby(['HADM_ID','ITEMID','AMOUNT','AMOUNTUOM','drug_length'])
df_inp = df_group.agg({
    'PATIENTWEIGHT': 'mean'
})
```

```python
df_inp
```

```python
df = pd.read_csv('merge_diag_pat_proc_icu_cpt_ser.csv')
df = df.merge(df_inp, how='inner', on='HADM_ID')
df.info()
df.to_csv('merge_diag_pat_proc_icu_cpt_ser_inp.csv',index=False)
```
